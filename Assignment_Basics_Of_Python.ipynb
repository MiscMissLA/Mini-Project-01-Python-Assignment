```python
# ques 1
L = [11,12,13,14]
L.append(50)
L.append(60)
print(L)
print("----")


if (11 in L) == True:
  try:
    del L[0]
  except:
    print("Not Present.")
else:
  print("Not Present.")

if (13 in L) == True:
  try:
    del L[1]
  except:
    print("Not Present.")
else:
  print("Not Present.")
print(L)
print("----")


L.sort()
print(L)
print("----")


L.sort(reverse = True)
print(L)
print("----")


if (13 in L) == True:
  print("Present.")
else:
  print("Not Present.")
print("----")


for i in L:
  print(i, " count:  ", L.count(i))
print("----")


sum = 0
for i in L:
  sum+=i
print("sum is: ", sum)
print("----")


sum_odd = 0
sum_even = 0
for i in L:
  if (i%2!=0):
    sum_odd+=i
  else:
    sum_even += i
print("sum of odd numbers: ", sum_odd)
print("----")


print("sum of even numbers: ", sum_even)
print("----")


import math as m
flag = 0
sum_prime = 0
for i in L:
  for j in range(2, int(m.sqrt(i))+1):
    if (i%j == 0):
        print("Not Prime")
        flag = 1
        break
  if (flag == 0):
    sum_prime += i
    print("Prime ", i)
print("sum of prime numbers in list: ", sum_prime)
print("----")


L.clear()
print(L)
print("----")


del L
try:
  print(L)
except:
  print("Deleted.")
print("----")

```

```python
# ques 3
S1 = {10, 20, 30, 40, 50, 60}
S2 = {40, 50, 60, 70, 80, 90}

print(S1)
print(S2)
print("----")

S1.add(55)
S1.add(66)
print(S1)
print("----")

S1.remove(10)
S1.remove(30)
print(S1)
print("----")

if (40 in S1):
  print("40 is Present.")
else:
  print("40 is Not Present.")
print("----")

print("Union of S1 and S2: ", S1.union(S2))
print("----")

print("Intersection of S1 and S2: ", S1.intersection(S2))
print("----")

print("Difference of S1 and S2: ", S1-S2, S1.difference(S2), " Symmetric Difference: ", S1.symmetric_difference(S2))
print("----")
```

```python
# ques 5
import random as r
L1 = []
L2 = []

print(L1)
print(L2)
print("----")

for i in range(1, 11):
  L1.append(r.randint(10,30))
  L2.append(r.randint(10,30))

print(L1)
print(L2)
print("----")

for i in L1:
  if i in L2:
    print(i, "is common in both L1 and L2.")
print("----")

L = []
for i in L1:
  if i not in L:
    L.append(i)
print("Unique elements in L1", L)

L.clear()

for i in L2:
  if i not in L:
    L.append(i)
print("Unique elements in L2", L)

L = set(L1)
print(L)

L = set(L2)
print(L)
print("----")

print("Minimum of L1", min(L1))
print("Minimum of L2", min(L2))
print("----")

print("Maximum of L1", max(L1))
print("Maximum of L2", max(L2))
print("----")

print("sum of L1", sum(L1))
print("sum of L2", sum(L2))
print("----")

```

```python
# ques 6
import random as r
import math as m
L = []
for i in range(1,101):
  L.append(r.randint(100,900))

count_odd = 0
count_even = 0
count_prime = 0

for i in L:
  flag = 0
  if i%2 == 0:
    count_even += 1
    print(i, "is even.")
  else:
    count_odd += 1
    print(i, "is odd.")

  for j in range(2, int(m.sqrt(i))+1):
    if i%j == 0:
      flag = 1
      break

  if flag == 0:
    print(i, "is prime.")
    count_prime += 1
  else:
    print(i, "is not prime.")

print("odd count: ", count_odd)
print("even count: ", count_even)
print("prime count: ", count_prime)
```

```python
# ques 13
n = int(input("Enter the number of students: "))
D = {}
for i in range(0, n):
  name = input("Enter the name of the student: ")
  sum = 0
  for j in range(0,5):
    s = "Enter the marks of the student in subject %d: " %(j+1)
    marks = int(input(s))
    sum += marks
  avg = sum/5;
  D[name] = avg
print(D)

for i in D:
  if D[i] == max(D.values()):
    print("Student with highest avg marks: ", i)
  if D[i] == min(D.values()):
    print("Student with lowest avg marks: ", i)





```
